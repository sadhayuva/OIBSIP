```python
# Import libraries
import pandas as pd
import numpy as np

# Step 1: Load dataset
data = pd.read_csv("CAvideos.csv")

# Display first rows
print("First 5 rows:")
print(data.head())

# Dataset info
print("\nDataset Info:")
print(data.info())

# ---------------------------------------------------
# Step 2: Check Missing Values
# ---------------------------------------------------
print("\nMissing values in each column:")
print(data.isnull().sum())

# Fill missing values for description
data['description'] = data['description'].fillna("No Description")

# ---------------------------------------------------
# Step 3: Remove Duplicate Records
# ---------------------------------------------------
print("\nNumber of duplicates:", data.duplicated().sum())

data = data.drop_duplicates()

print("Duplicates removed")

# ---------------------------------------------------
# Step 4: Standardize Column Names
# ---------------------------------------------------
data.columns = data.columns.str.lower().str.replace(" ", "_")

print("\nStandardized column names:")
print(data.columns)

# ---------------------------------------------------
# Step 5: Convert Date Format
# ---------------------------------------------------
data['publish_time'] = pd.to_datetime(data['publish_time'])

# ---------------------------------------------------
# Step 6: Outlier Detection (Using IQR)
# Example: views column
# ---------------------------------------------------

Q1 = data['views'].quantile(0.25)
Q3 = data['views'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = data[(data['views'] < lower) | (data['views'] > upper)]

print("\nNumber of Outliers in Views:", len(outliers))

# Optionally remove them
data_cleaned = data[(data['views'] >= lower) & (data['views'] <= upper)]

# ---------------------------------------------------
# Step 7: Save Cleaned Dataset
# ---------------------------------------------------
data_cleaned.to_csv("cleaned_youtube_data.csv", index=False)

print("\nCleaned dataset saved successfully!")
```

```python
# Import libraries
import pandas as pd
import numpy as np

# ---------------------------------------------------
# Step 1: Load Dataset
# ---------------------------------------------------
data = pd.read_csv("AB_NYC_2019.csv")

print("First 5 rows:")
print(data.head())

print("\nDataset Shape:")
print(data.shape)

print("\nDataset Info:")
print(data.info())

# ---------------------------------------------------
# Step 2: Check Missing Values
# ---------------------------------------------------
print("\nMissing Values:")
print(data.isnull().sum())

# Fill missing values
data['name'] = data['name'].fillna("Unknown")
data['host_name'] = data['host_name'].fillna("Unknown")

# Fill numeric missing values with median
data['reviews_per_month'] = data['reviews_per_month'].fillna(data['reviews_per_month'].median())

# ---------------------------------------------------
# Step 3: Remove Duplicate Records
# ---------------------------------------------------
print("\nDuplicate rows:", data.duplicated().sum())

data = data.drop_duplicates()

print("Duplicates removed!")

# ---------------------------------------------------
# Step 4: Standardize Column Names
# ---------------------------------------------------
data.columns = data.columns.str.lower().str.replace(" ", "_")

print("\nStandardized Columns:")
print(data.columns)

# ---------------------------------------------------
# Step 5: Convert Date Column
# ---------------------------------------------------
data['last_review'] = pd.to_datetime(data['last_review'], errors='coerce')

# ---------------------------------------------------
# Step 6: Detect Outliers (Using IQR)
# Example: price column
# ---------------------------------------------------
Q1 = data['price'].quantile(0.25)
Q3 = data['price'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = data[(data['price'] < lower) | (data['price'] > upper)]

print("\nNumber of price outliers:", len(outliers))

# Remove outliers
data_cleaned = data[(data['price'] >= lower) & (data['price'] <= upper)]

# ---------------------------------------------------
# Step 7: Standardization Example
# Convert room_type to lowercase
# ---------------------------------------------------
data_cleaned['room_type'] = data_cleaned['room_type'].str.lower()

# ---------------------------------------------------
# Step 8: Save Cleaned Dataset
# ---------------------------------------------------
data_cleaned.to_csv("cleaned_airbnb_data.csv", index=False)

print("\nCleaned dataset saved successfully!")
```
